```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
files=[file for file in os.listdir("./Sales datasets")]
all_data=pd.DataFrame()
for file in files:
    sales_data=pd.read_csv("./Sales datasets/"+file)
    all_data=pd.concat([all_data,sales_data])
all_data.to_csv("all_data.csv",index=False)
```

```python
allData=pd.read_csv("all_data.csv")
allData
```

Find null values

```python
allData.isna().sum()
```

```python
allData[allData.isna().any(axis=1)]
```

Drop null Values

```python
allData.dropna(inplace=True)
allData
```

Find duplicated values

```python
allData.drop_duplicates(inplace=True)
allData
```

```python
allData.drop(allData[allData["Order ID"]=="Order ID"].index,inplace=True)
allData
```

# QUESTION 1:WHAT was the best month for sales? How much was earned that month?

1.Add month column

```python
allData["Month"]=allData["Order Date"].str[0:2]
allData["Month"]=allData["Month"].astype('int32')
```

2.Add sales column

```python
allData["Quantity Ordered"]=allData["Quantity Ordered"].astype('int32')
allData["Price Each"]=allData["Price Each"].astype('float64')
allData["Sales"]=allData["Price Each"]*allData["Quantity Ordered"]
allData
```

3.Find best month

```python
sales_results=allData.groupby("Month").sum()
sales_results.sort_values(by="Sales",ascending=False)
```

```python
months=range(1,13)
plt.bar(months,sales_results["Sales"])
plt.xticks(months)
plt.xlabel("months")
plt.ylabel("Sales in USD")
plt.show()
```

# Question 2.what city sold the most product?

```python
allData["City"]=allData["Purchase Address"].str.split(",", expand = True)[1]
allData
```

```python
city_results=allData.groupby("City").sum()
city_results.sort_values(by="Sales",ascending=False)
city_results
```

```python

plt.bar(city_results.index,city_results["Sales"])
plt.xticks(city_results.index,rotation="vertical")
plt.xlabel("City")
plt.ylabel("Sales in USD($)")
plt.show()
```

# Question 3. what time should we display advertisements?

```python
allData["Hour"]=pd.to_datetime(allData["Order Date"]).dt.hour
allData_hour_results=allData.groupby("Hour").count()
plt.plot(allData_hour_results.index,allData_hour_results,marker="o",mec="k",ms=12)
plt.xticks(allData_hour_results.index)
plt.grid()
plt.show()
```

# question 4.which products are often sold together?

```python
allData_grouped=allData[allData["Order ID"].duplicated()]
allData_grouped["Grouped_products"]=allData_grouped.groupby("Order ID")["Product"].transform(lambda x:",".join(x))
allData_grouped=allData_grouped[["Order ID","Grouped_products"]].drop_duplicates()
allData_grouped
```

```python
from itertools import combinations
from collections import Counter
count=Counter()
for row in allData_grouped["Grouped_products"]:
    count_list=row.split(",")
    count.update(Counter(combinations(count_list,2)))
count.most_common(10)
```

# Question 5:what product sold the most?

```python
allData_most_sold=allData.groupby("Product").sum()
plt.bar(allData_most_sold.index,allData_most_sold["Quantity Ordered"])
plt.xticks(allData_most_sold.index,rotation=90)
plt.show()
allData_most_sold
```

```python
allData_mean_price=allData.groupby("Product").mean()
```

```python
fig,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.bar(allData_most_sold.index,allData_most_sold["Quantity Ordered"])
ax2.plot(allData_most_sold.index,allData_mean_price,marker=".",mec="k",ms=12)
ax1.set_xticklabels(allData_most_sold.index, rotation=90)
ax1.set_xlabel('Products')
ax1.set_ylabel('Sales', color='g')
ax2.set_ylabel('Price Each', color='b')
plt.show()
```
